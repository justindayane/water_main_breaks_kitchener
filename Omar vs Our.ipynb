## Visuals of metrics from our approach, Omar et al report, and Omar et al replication without COS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading the metrics from the csv file into the df
our_metrics = pd.read_csv("//Users//justindayane//Documents//Spring_24//Supervised_Research//Neh//our_metrics_Apr21.csv")
our_metrics['Model'] = ['DT', 'GNB', 'KNN', 'LoR', 'RF', 'XGB'] 

#reading the replicated metrics from the csv file into the df
replicated_metrics = pd.read_csv("//Users//justindayane//Documents//Spring_24//Supervised_Research//Neh//omar_replicated_metrics_21Apr.csv")
replicated_metrics['Model'] = our_metrics['Model']

#reading the reported metrics from the the csv file compiled from the paper. Keep in mind that XGB was not done in the paper
#Also keep in mind that only Accuracy had its standard deviation reported
reported_metrics = pd.read_csv("//Users//justindayane//Documents//Spring_24//Supervised_Research//Neh//omar_reported_metrics.csv")
reported_metrics['Model'] = ['DT', 'GNB', 'KNN', 'LoR', 'RF', 'XGB'] # one less metric XGB (not in paper)

#display(reported_metrics)
```

```python
list(replicated_metrics['Model'])
```

```python
#Focus on the accuracy part
our_acc = list(our_metrics['Accuracy'])
reported_acc = list(reported_metrics['Accuracy'])
replicated_acc = list(replicated_metrics['Accuracy'])

#Focus on the standard Deviation
our_acc_std = list(our_metrics['Accuracy_std'])
reported_acc_std = list(reported_metrics['Accuracy_std'])
replicated_acc_std = list(replicated_metrics['Accuracy_std'])

display(replicated_acc)
```

```python
#Plotting
barWidth = 0.25
fig = plt.subplots(figsize = (10,6))

br1 = [x for x in np.arange(len(our_acc))]
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, our_acc, color = 'red', width = barWidth, edgecolor = 'black', label = 'Ours')
plt.errorbar(br1, our_acc, yerr=our_acc_std, fmt='none', capsize = 5, ecolor='black')

plt.bar(br2, replicated_acc, color = 'green', width = barWidth, edgecolor = 'black', label = 'Omar et al')
plt.errorbar(br2, replicated_acc, yerr=replicated_acc_std, fmt='none', capsize = 5, ecolor='black')

#plt.bar(br3, reported_acc, color = 'blue', width = barWidth, edgecolor = 'black', label = 'Reported Omar et al')
#plt.errorbar(br3, reported_acc, yerr=reported_acc_std, fmt='none', capsize = 5, ecolor='black')

plt.xlabel('Model', fontweight ='bold', fontsize = 15)
plt.ylabel('Accuracy', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(our_acc))], ['Decision Tree', 'GNB', 'KNN', 'Logistic Regression', 'Random forest', 'XGBoost'])

plt.legend()
#plt.savefig("Images/models_acc_comp.png")
plt.show()
```
